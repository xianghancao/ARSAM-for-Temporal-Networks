```python

path = geopandas.datasets.get_path("naturalearth_lowres")
df = geopandas.read_file(path)
cm=1/2.54
ax = df.plot(figsize=(20, 20), color='grey', alpha=0.7, )

plt.rcParams.update({
    "text.usetex": True,
    "font.family": "Helvetica",
    "font.size":10
})



#ax.set_title('Europe Map')

df = list_df#[list_df['select']=='v']
jw = df['经纬度']
city = df['城市英文名']
for j, i in enumerate(jw):
    lon = float(i.split('经纬度：')[1].split(' , ')[0])
    lat = float(i.split('经纬度：')[1].split(' , ')[1])
    ax.plot(lon, lat, 's',
            color='red',
           label=city.iloc[j])
    if df['区域'].iloc[j] != '欧洲':
        plt.text(lon+1.5, lat, city.iloc[j],
               horizontalalignment='left')
#plt.xlim((-15,30))
#plt.ylim((30,70))
plt.xticks([])
plt.yticks([])
plt.savefig('image/whole_map.pdf')
```

```python

path = geopandas.datasets.get_path("naturalearth_lowres")
df = geopandas.read_file(path)

ax = df.plot(figsize=(8,8), color='grey', alpha=0.7)

ax.set_title('Europe Map')

df = list_df#[list_df['select']=='v']
df = df[df['区域'] == '欧洲']
jw = df['经纬度']
city = df['城市英文名']
for j, i in enumerate(jw):
    lon = float(i.split('经纬度：')[1].split(' , ')[0])
    lat = float(i.split('经纬度：')[1].split(' , ')[1])
    ax.plot(lon, lat, 's',
            color='red',
           label=city.iloc[j])
    plt.text(lon+1, lat, city.iloc[j],
            horizontalalignment='left')
plt.xlim((-15,30))
plt.ylim((30,70))
plt.xticks([])
plt.yticks([])
plt.savefig('image/map-Europe.pdf')
```
